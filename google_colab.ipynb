```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
credited_card_df = pd.read_csv('/content/creditcard.csv', on_bad_lines='skip', low_memory=False)

credited_card_df.head()
```

```python
credited_card_df.info()
```

```python
credited_card_df.describe()
```

```python
legit = credited_card_df[credited_card_df.Class==0]
fraud = credited_card_df[credited_card_df['Class']==1]
```

```python
fraud.shape
```

```python
legit.shape
```

```python
legit_sample = legit.sample(n=145)
credited_card_df = pd.concat([legit_sample,fraud],axis=0)
```

```python
credited_card_df['Class'].value_counts()
```

```python
print(credited_card_df.dtypes)
```

```python
for col in credited_card_df.columns:
    credited_card_df[col] = pd.to_numeric(credited_card_df[col], errors='coerce')
credited_card_df.fillna(0, inplace=True)
```

```python
numeric_cols = credited_card_df.select_dtypes(include=[np.number]).columns.tolist()
mean_values = credited_card_df[numeric_cols].groupby(credited_card_df['Class']).mean()
print(mean_values)
```

```python
credited_card_df.fillna(0, inplace=True)  # or use credited_card_df.dropna(inplace=True)
```

```python
credited_card_df.groupby('Class').mean()
```

```python
X = credited_card_df.drop('Class', axis=1)  # Features
Y = credited_card_df['Class']  # Target variable
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LogisticRegression(max_iter=300, solver='liblinear')  # Adjusted parameters
```

```python
model.fit(X_train, Y_train)
```

```python
ypred = model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, ypred)
print(f'Accuracy: {accuracy:.2f}')
```
